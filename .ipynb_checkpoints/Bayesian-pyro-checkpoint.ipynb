## Simple example of the Stochastic Variational inference (SVI)

```python
from __future__ import print_function
import math
import os
import torch
import torch.distributions.constraints as constraints
import pyro
from pyro.optim import Adam
from pyro.infer import SVI, Trace_ELBO
import pyro.distributions as dist

pyro.enable_validation(True)
```

```python
from scipy.stats import multivariate_normal

# define helper function (spot with amplitude A)
def classA(A, B):
    N = 9
    sigma = 2
    mu = (4, 4)
    x, y = np.mgrid[0:N:1, 0:N:1]
    pos = np.dstack((x, y))
    rv = multivariate_normal([mu[0], mu[1]], cov=sigma*np.diag(np.ones((2,))))
    return A*2*np.pi*sigma*torch.tensor(rv.pdf(pos).ravel()).float() + B

def classB(B):
    return torch.ones(81)*B
```

```python
# simulate data
data = []
for _ in range(10):
    a_sim = pyro.sample('a_sim', dist.Normal(130, 5)) # true distribution
    #print(a_sim.item())
    with pyro.plate("data", size=81):
        img_sim = pyro.sample('img_sim', dist.Normal(classA(a_sim, 400.), 10.))
        data.append(img_sim) # likelihood

plt.figure(figsize=(15,6))
plt.subplot(2,5,1)
plt.imshow(img_sim.reshape(9,9), cmap='gray')
for i in range(1,10):
    plt.subplot(2,5,i+1)
    plt.imshow((data[i]).reshape(9,9), cmap='gray')

plt.show()
```

```python
# clear the param store in case we're in a REPL
pyro.clear_param_store()

def model(data):
    # define the hyperparameters that control the beta prior
    alpha0 = torch.tensor(150.)
    beta0 = torch.tensor(10.0)
    # loop over the observed data
    for i in range(len(data)):
        f = pyro.sample("f_{}".format(i), dist.Normal(alpha0, beta0))
        with pyro.plate("observe_data"):
            # observe datapoint i using the normal likelihood
            pyro.sample("obs_{}".format(i), dist.Normal(classA(f), 10.), obs=data[i])

def guide(data):
    # variational distribution
    alpha_q = pyro.param("alpha_q", torch.tensor(140.))
    beta_q = pyro.param("beta_q", torch.tensor(10.0),
                        constraint=constraints.positive)
    # sample latent_fairness from the distribution Beta(alpha_q, beta_q)
    for i in range(len(data)):
        f = pyro.sample("f_{}".format(i), dist.Normal(alpha_q, beta_q))

# setup the optimizer
adam_params = {"lr": 0.03, "betas": (0.90, 0.999)}
optimizer = Adam(adam_params)

# setup the inference algorithm
svi = SVI(model, guide, optimizer, loss=Trace_ELBO())

losses, a,b  = [], [], []
n_steps = 2000
# do gradient steps
for step in range(n_steps):
    losses.append(svi.step(data))
    a.append(pyro.param("alpha_q").item())
    b.append(pyro.param("beta_q").item())
    if step % 100 == 0:
        print('.', end='')

# grab the learned variational parameters
alpha_q = pyro.param("alpha_q").item()
beta_q = pyro.param("beta_q").item()

print("\nposterior height {} and width {}".format(alpha_q, beta_q))
```

```python
plt.subplot(1,2,1)
plt.ylabel('a')
plt.plot([0,n_steps],[130,130], 'k:')
plt.plot(a)

plt.subplot(1,2,2)
plt.ylabel('b')
plt.plot([0,n_steps],[5,5], 'k:')
plt.plot(b)
plt.tight_layout()
```

## Mixture model

```python
# simulate data
K = 2
data = []
for _ in range(10):
    assignment = pyro.sample('assignment', dist.Categorical(torch.tensor([0.3,0.7])))
    a_sim = pyro.sample('a_sim', dist.Normal(130, 5)) # true distribution
    #print(a_sim.item())
    with pyro.plate("data", size=81):
        if assignment.item() == 1:
            img_sim = pyro.sample('img_sim', dist.Normal(classA(a_sim, 400.), 10.))
        else:
            img_sim = pyro.sample('img_sim', dist.Normal(classB(400.), 10.))
        data.append(img_sim) # likelihood

plt.figure(figsize=(15,6))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow((data[i]).reshape(9,9), cmap='gray')

plt.show()
```

```python
# clear the param store in case we're in a REPL
pyro.clear_param_store()

def model(data):
    # define the hyperparameters that control the beta prior
    alpha0 = torch.tensor(150.)
    beta0 = torch.tensor(10.0)
    # loop over the observed data
    weights = pyro.sample('weights', dist.Dirichlet(0.5 * torch.ones(K)))
    for i in range(len(data)):
        z = pyro.sample('z_{}'.format(i), dist.Categorical(weights))
        f = pyro.sample("f_{}".format(i), dist.Normal(alpha0, beta0))
        with pyro.plate("observe_data"):
            # observe datapoint i using the normal likelihood
            if z.item() == 1:
                
                pyro.sample("obs_{}".format(i), dist.Normal(classA(f, 400.), 10.), obs=data[i])
            else:
                pyro.sample("obs_{}".format(i), dist.Normal(classB(400.), 10.), obs=data[i])

def guide(data):
    # variational distribution
    alpha_q = pyro.param("alpha_q", torch.tensor(140.))
    beta_q = pyro.param("beta_q", torch.tensor(10.0),
                        constraint=constraints.positive)
    w_a = pyro.param("w_a", 0.5 * torch.ones(K), constraint=constraints.simplex)
    weights = pyro.sample('weights', dist.Dirichlet(w_a))
    # sample latent_fairness from the distribution Beta(alpha_q, beta_q)
    z_list = []
    for i in range(len(data)):
        z = pyro.sample('z_{}'.format(i), dist.Categorical(weights))
        f = pyro.sample("f_{}".format(i), dist.Normal(alpha_q, beta_q))
        z_list.append(z)
    return z_list

# setup the optimizer
adam_params = {"lr": 0.03, "betas": (0.90, 0.999)}
optimizer = Adam(adam_params)

# setup the inference algorithm
svi = SVI(model, guide, optimizer, loss=Trace_ELBO())

losses, a,b  = [], [], []
n_steps = 2000
# do gradient steps
for step in range(n_steps):
    losses.append(svi.step(data))
    a.append(pyro.param("alpha_q").item())
    b.append(pyro.param("beta_q").item())
    if step % 100 == 0:
        print('.', end='')

# grab the learned variational parameters
alpha_q = pyro.param("alpha_q").item()
beta_q = pyro.param("beta_q").item()
w_a = pyro.param("w_a")

print("\nposterior height {} and width {}".format(alpha_q, beta_q))
print(w_a)
```

```python
plt.subplot(1,2,1)
plt.ylabel('a')
plt.plot([0,n_steps],[130,130], 'k:')
plt.plot(a)

plt.subplot(1,2,2)
plt.ylabel('b')
plt.plot([0,n_steps],[5,5], 'k:')
plt.plot(b)
plt.tight_layout()
```

## GMM Tutorial

```python
import os
from collections import defaultdict
import numpy as np
import scipy.stats
import torch
from torch.distributions import constraints
from matplotlib import pyplot
%matplotlib inline

import pyro
import pyro.distributions as dist
from pyro import poutine
from pyro.contrib.autoguide import AutoDelta
from pyro.optim import Adam
from pyro.infer import SVI, TraceEnum_ELBO, config_enumerate, infer_discrete

pyro.enable_validation(True)
```

```python
K = 2  # Fixed number of components.

@config_enumerate
def model(data):
    # Global variables.
    weights = pyro.sample('weights', dist.Dirichlet(0.5 * torch.ones(K)))
    #scale = pyro.sample('scale', dist.LogNormal(0., 2.))
    scale = pyro.sample('scale', dist.MultivariateNormal(torch.zeros(2), torch.eye(2)*2))
    scale = torch.exp(scale)
    with pyro.plate('components', K):
        locs = pyro.sample('locs', dist.MultivariateNormal(torch.zeros(2), torch.eye(2)))
    
    for i in range(len(data)):
        with pyro.plate('data'):
            # Local variables.
            assignment = pyro.sample('assignment_{}'.format(i), dist.Categorical(weights))
            pyro.sample('obs_{}'.format(i), dist.MultivariateNormal(locs[assignment], scale), obs=data[i])

global_guide = AutoDelta(poutine.block(model, expose=['weights', 'locs', 'scale']))
```

```python
optim = pyro.optim.Adam({'lr': 0.1, 'betas': [0.8, 0.99]})
elbo = TraceEnum_ELBO(max_plate_nesting=1)
svi = SVI(model, global_guide, optim, loss=elbo)
```

```python
var_x = (torch.tensor([d[0] for d in data]).var() / 2).sqrt()
var_y = (torch.tensor([d[1] for d in data]).var() / 2).sqrt()
torch.tensor([[var_x, 0.], [var_y, 0.]])
```

```python
def initialize(seed):
    pyro.set_rng_seed(seed)
    pyro.clear_param_store()
    # Initialize weights to uniform.
    pyro.param('auto_weights', 0.5 * torch.ones(K), constraint=constraints.simplex)
    # Assume half of the data variance is due to intra-component noise.
    pyro.param('auto_scale', torch.tensor([[var_x, 0.], [0., var_y]]), constraint=constraints.positive)
    # Initialize means from a subsample of data.
    pyro.param('auto_locs', [data[i.item()] for i in torch.multinomial(torch.ones(len(data)) / len(data), K).numpy()]);
    loss = svi.loss(model, global_guide, data)
    return loss

# Choose the best among 100 random initializations.
loss, seed = min((initialize(seed), seed) for seed in range(100))
initialize(seed)
print('seed = {}, initial_loss = {}'.format(seed, loss))
```

```python
# Register hooks to monitor gradient norms.
gradient_norms = defaultdict(list)
for name, value in pyro.get_param_store().named_parameters():
    value.register_hook(lambda g, name=name: gradient_norms[name].append(g.norm().item()))

losses = []
for i in range(200):
    loss = svi.step(data)
    losses.append(loss)
    print('.' if i % 100 else '\n', end='')
```

```python
pyplot.figure(figsize=(10,3), dpi=100).set_facecolor('white')
pyplot.plot(losses)
pyplot.xlabel('iters')
pyplot.ylabel('loss')
pyplot.yscale('log')
pyplot.title('Convergence of SVI');
```

```python
map_estimates = global_guide(data)
weights = map_estimates['weights']
locs = map_estimates['locs']
scale = map_estimates['scale']
print('weights = {}'.format(weights.data.numpy()))
print('locs = {}'.format(locs.data.numpy()))
print('scale = {}'.format(scale.data.numpy()))
```

```python
X = np.arange(-3,15,0.1)
Y1 = weights[0].item() * scipy.stats.norm.pdf((X - locs[0].item()) / scale.item())
Y2 = weights[1].item() * scipy.stats.norm.pdf((X - locs[1].item()) / scale.item())

pyplot.figure(figsize=(10, 4), dpi=100).set_facecolor('white')
pyplot.plot(X, Y1, 'r-')
pyplot.plot(X, Y2, 'b-')
pyplot.plot(X, Y1 + Y2, 'k--')
pyplot.plot(data.data.numpy(), np.zeros(len(data)), 'k*')
pyplot.title('Density of two-component mixture model')
pyplot.ylabel('probability density');
```
