```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib tk

def surface_plot (matrix, ax, **kwargs):
    # acquire the cartesian coordinate matrices from the matrix
    # x is cols, y is rows
    (x, y) = np.meshgrid(np.arange(matrix.shape[0]), np.arange(matrix.shape[1]))
    
    
    surf = ax.plot_surface(x, y, matrix, **kwargs)
    return surf

def wireframe_plot (matrix, ax, **kwargs):
    # acquire the cartesian coordinate matrices from the matrix
    # x is cols, y is rows
    (x, y) = np.meshgrid(np.arange(matrix.shape[0]), np.arange(matrix.shape[1]))
    
    
    surf = ax.plot_wireframe(x, y, matrix, **kwargs)
    return surf

def scatter_plot (matrix, ax, **kwargs):
    # acquire the cartesian coordinate matrices from the matrix
    # x is cols, y is rows
    (x, y) = np.meshgrid(np.arange(matrix.shape[0]), np.arange(matrix.shape[1]))
    
    
    surf = ax.scatter(x, y, matrix, **kwargs)
    return surf


fig = plt.figure()
ax = fig.add_subplot(221, projection='3d')
surf = surface_plot(classA(100,2,220,10).cpu().numpy(), ax, cmap=plt.cm.coolwarm)
surf = scatter_plot(smd["data"][0,97].cpu().numpy(), ax, cmap=plt.cm.coolwarm)
#fig.colorbar(surf)

ax.set_xlabel('X (cols)')
ax.set_ylabel('Y (rows)')
ax.set_zlabel('Z (values)')

ax = fig.add_subplot(222, projection='3d')
surf = surface_plot(classB(400,10).cpu().numpy(), ax, cmap=plt.cm.coolwarm)

plt.show()
```

### Class membership probabilities

```python
@config_enumerate
def full_guide(data, K):
    N, D, _ = data.shape
    # Global variables.
    with poutine.block(hide_types=["param"]):  # Keep our learned values of global parameters.
        guide(data, K)
    # Local variables.
    with pyro.plate("sample_size", N):
        z_probs = pyro.param("z_probs", torch.ones(N, K) / K, constraint=constraints.simplex)
        pyro.sample("z", dist.Categorical(z_probs))
```

```python
optim = pyro.optim.Adam({'lr': 0.2, 'betas': [0.8, 0.99]})
elbo = TraceEnum_ELBO()
svi = SVI(model, full_guide, optim, loss=elbo)

losses, zs = [], []
for i in range(200):
    loss = svi.step(smd["data"][0], K)
    losses.append(loss)
    zs.append(pyro.param("z_probs"))
    print('.' if i % 100 else '\n', end='')
```

# Bayesian analysis

### Variables

$X = \{x_{n}\}$ (N images)

$ x_n = \{x_{n,d}\} $ (D pixels; D = 9x9 = 81)

$\pi = \{\pi_{k}\} = [\pi_0, \pi_1]$ (K classes; K = 2)

$\mu = \{ \mu_{k} \} = [\mu_{0}, \mu_{1}]$ (K classes)

$ \mu_k = \{\mu_{k,d}\} $ (D pixels)

$\mu_{0} = classB(b) $ (b - background level)

$\mu_{1} = classA(h, w, b) $ (h, w - height and width of the 2D Gaussian spot)

$\sigma$ is the noise in the image

### Bayes' theorem

<center>$$ p(\lambda | X) = \dfrac{p(X | \lambda) p(\lambda)}{p(X)} $$</center>

Marginal likelihood of one image:
<center>$$p (x_n|\lambda_n) = \sum_{k}^{K} \pi_{k} Poisson(x_n|\lambda_{k,n}) $$</center>

where:
<center>$$ Poisson(x_n|\lambda_{k,n}) = \prod_d^D Poisson(x_{n,d}|\lambda_{k,n,d}) $$</center>

Marginal likelihood of N images:
<center>$$p(X|\lambda) = \prod_{n}^{N} \bigg[ \sum_{k}^K \pi_{k} \prod_d^D Poisson(x_{n,d}|\lambda_{k,n,d}) \bigg]$$</center>

Prior:
<center>$$p (\lambda) = \prod_{k}^{K} \prod_{d}^{D} Gamma(\alpha_{k,d}(h,w,b)|\beta_{k,d}(h,w,b)) $$</center>

Posterior:
<center>$$p (\lambda | X) \simeq q(\lambda) = \prod_{k}^{K} \prod_{d}^{D} Gamma(\alpha_{k,d}, \beta_{k,d}) $$</center>

Constraint:
<center>$$ \dfrac{\alpha_{k,d}}{\beta_{k,d}} = class(k, h, w, b) $$</center>

<center>$$ p(\pi, \mu, \sigma | X) = \dfrac{p(X | \pi, \mu, \sigma) p(\pi, \mu, \sigma)}{p(X)}
= \dfrac{p(X | \pi, \mu, \sigma) p(\pi) p(\mu) p(\sigma)}{p(X)} $$</center>

Marginal likelihood of one image:
<center>$$p (x_n|\pi, \mu, \sigma) = \sum_k^K p(z=k) p(x_n|\mu_k, \sigma) = \sum_{k}^{K} \pi_{k} \mathcal{N}(x_n|\mu_{k},\sigma) $$</center>

where:
<center>$$ \mathcal{N}(x_n|\mu_{k},\sigma) = \prod_d^D \mathcal{N}(x_{n,d}|\mu_{k,d},\sigma) $$</center>

Marginal likelihood of N images:
<center>$$p(X|\pi, \mu, \sigma) = \prod_{n}^{N} \bigg[ \sum_{k}^K \pi_{k} \mathcal{N}(x_{n}|\mu_{k},\sigma) \bigg]$$</center>

### Priors

$ p(\pi) = Dir(\alpha_0) $

$ p(\mu) = p(h, w, b) = p(h)p(w)p(b) $

$ p(h) = HalfNormal(500) $

$ p(w) = Gamma(1, 0.1) $

$ p(b) = HalfNormal(1000) $

$ p(\sigma) = Gamma(1, 0.001) $

### Guides (approximations to posterior distributions)

$ p(\pi) = Dir(\pi\_conc) $

$ p(\mu) = p(h, w, b) = p(h)p(w)p(b) $

$ p(h) = \mathcal{N}(h\_loc, h\_scale) $

$ p(w^2) =  InverseGamma(w\_alpha, w\_beta)$

$ p(b) = \mathcal{N}(b\_loc, b\_scale) $

$ p(\sigma^2) =  InverseGamma(\sigma\_alpha, \sigma\_beta)$

```python
def delta_guide(data, N, K, sample_size):
    auto_weights = pyro.param("auto_weights", torch.ones(2)*0.5, constraint=constraints.simplex)
    weights = pyro.sample("weights", dist.Delta(auto_weights, event_dim=1))
    auto_height = pyro.param("auto_height", torch.tensor(100.))
    height = pyro.sample('height', dist.Delta(auto_height)) # true distribution
    auto_background = pyro.param("auto_background", torch.tensor(360.))
    background = pyro.sample('background', dist.Delta(auto_background))
    return {"weights": weights, "height": height, "background": background}
```
